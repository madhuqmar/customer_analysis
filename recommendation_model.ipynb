## Content-Based Recommendation Model

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df4 = pd.read_csv("data/Tickets_14Nov23_4pm.csv", encoding='ISO-8859-1', low_memory=False)
df4.head()
```

```python
len(df4)
```

```python
# Try to convert the 'column_name' to int64, handling errors
try:
    df4['TicketID'] = df4['TicketID'].astype('int64')
except ValueError:
    # Handle non-numeric values, for example, replacing them with NaN
    df4['TicketID'] = pd.to_numeric(df4['TicketID'], errors='coerce')

#print(df4.dtypes)
```

```python
df3 = pd.read_csv("data/Ticket_Product_Details_14Nov23_4pm.csv", encoding='ISO-8859-1', low_memory=False)
df3.head()
```

```python
len(df3)
```

```python
client_services = pd.merge(df3, df4, on='TicketID', how='left')
client_services.head()
```

```python
len(client_services)
```

```python
client_services['Frequency'] = client_services.groupby(['ClientID', 'Descr'])['ClientID'].transform('size')

# Reset index to make 'SalonID' a regular column
client_services.reset_index(inplace=True)

# Print the resulting DataFrame
client_services.head()
```

```python
len(client_services)
```

```python
client_services = client_services[client_services['Descr'] != '35']
```

```python
client_services = client_services[client_services['Descr'] != '10']
```

```python
client_services = client_services[client_services['Descr'] != '0']
```

```python
len(client_services)
```

```python
client_services = client_services.drop_duplicates()
```

```python
len(client_services)
```

```python
len(df4['ClientID'].unique())
```

```python
client_services = client_services[['ClientID', 'Descr', 'Frequency']]
```

```python
client_services.sort_values(by='Frequency', ascending=False)
```

```python
pivot_df = client_services.pivot_table(index='ClientID', columns='Descr', values='Frequency', fill_value=0)
pivot_df
```

```python
# Calculate cosine similarity between customers
customer_similarity = cosine_similarity(pivot_df)

# Generate recommendations for each customer
recommendations = {}
for idx, customer in enumerate(pivot_df.index):
    similar_customers = sorted(list(enumerate(customer_similarity[idx])), key=lambda x: x[1], reverse=True)
    recommendations[customer] = []
    for similar_customer, similarity_score in similar_customers[1:]:  # Exclude the customer itself
        similar_products = pivot_df.columns[pivot_df.iloc[similar_customer] > 0].tolist()
        for product in similar_products:
            if pivot_df.loc[customer, product] == 0:  # Check if the customer hasn't bought the product yet
                recommendations[customer].append(product)
```

```python
# Print recommendations for each customer
for customer, recommended_products in recommendations.items():
    print(f"Customer {customer}: {recommended_products}")
```
